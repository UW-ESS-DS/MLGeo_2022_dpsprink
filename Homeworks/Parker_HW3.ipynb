### a) Simple linear regression (2 points)
We will now perform a regression between the parameters (or their log!). Linear regressions are models that can be imported from scikit-learn. Log/exp functions in numpy as ``np.log()`` and ``np.exp()``.
Remember that a linear regression is finding $a$ and $b$ knowing both $x$ and the data $y$ in $y = Ax +b$. We want to predict ice thickness from a crude estimate of the glacier area.

__Tips__: 
a. make sure that the dimensions are correct and that there is no NaNs and zeros.
b. Make sure to inport the scikit learn linear regression function and the error metrics.

```python
# solution
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# convert the data into numpy arrays.
th = np.asarray(df['mean_thickness']).reshape(-1, 1)# reshaping was necessary to be an argument of Linear regress
ar = np.asarray(np.log(df['area'])).reshape(-1, 1)


# perform the linear regression. First we will use the entire available data
regr = LinearRegression()
# we will first perform the fit:
regr.fit(ar,th)
# We will first predict the fit:
Tpred=regr.predict(ar) 


print(f'The mean thickness is: {Tpred.mean()} (m)')

# The mean squared error
print('Mean squared error (m): %.2f'
      % mean_squared_error(th, Tpred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(th, Tpred))
```

Make a plot of the data and the linear regression your performed

```python
# solution
plt.figure()
plt.plot(ar,th,'.k',label='observed');ax.grid(True);ax.set_ylabel('Mean_Thickness (m)')
plt.plot(ar,Tpred,'.r',label='predicted')
plt.grid(True)
plt.xticks()
plt.yticks()
plt.xlabel('Area (km^2)')
plt.ylabel('Mean_thickness (m)')
# plt.xscale('log')
plt.legend(loc='upper left')
plt.show()
```

### b) Leave One Out Cross Validation linear regression (1 point)


Perform the LOCCV on the ``area`` and ``thickness`` values. Predict the ``thickness`` value knowing a ``area`` value. Use material seen in class. Make a plot of your fit.

```python
from sklearn.model_selection import LeaveOneOut
# solution
loo = LeaveOneOut()
thpred = np.zeros(len(th)) # initalize a vector to store the regression values
mse_train = np.zeros(len(th))
mse_val = np.zeros(len(th))
r2s = np.zeros(len(th))
i=0
for train_index, test_index in loo.split(th):    
    th_train, th_val = th[train_index], th[test_index]
    ar_train, ar_val = ar[train_index], ar[test_index]
    # now fit the data on the training set.
    regr = LinearRegression()
    # Fit on training data:
    regr.fit(ar_train,th_train)
    # We will first predict the fit:
    thpred_train=regr.predict(ar_train) 
    thpred_val=regr.predict(ar_val) 

    # The coefficients
    # thk[i]= regr.coef_[0][0]
    mse_train[i]= mean_squared_error(th_train, thpred_train)
    mse_val[i]= mean_squared_error(th_val, thpred_val)
    # r2s[i]=r2_score(th_val, thpred_val)
    thpred[i]=thpred_val
    i+=1
    
# the data shows cleary a trend, so the predictions of the trends are close to each other:
print(f'mean of the thickness estimates {np.mean(thpred):.2f} and the standard deviation {np.std(thpred):.2f}')
# the test error is the average of the mean-square-errors
print("CV = %4.2f"%(np.mean(mse_val)))
```

```python
plt.figure()
plt.plot(ar,th,'.k',label='observed')
plt.plot(ar,thpred,'.r',label='LOO predicted')
plt.xlabel('area km^2')
plt.ylabel('Predicted thickness (m)')
plt.legend()
```
